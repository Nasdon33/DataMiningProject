```python
import pandas as pd
import numpy as np

# The dataset was created with the notebook "DatasetCreation".

routes = pd.read_csv('route_trips.csv',delimiter=',',encoding='utf-8')
routes.drop(routes.columns[0], axis=1, inplace = True)
```

```python
routes
```

```python
import re
delimiters = ",", "{", "}", "'"
regexPattern = '|'.join(map(re.escape, delimiters))
```

```python
r = []
for i in range(0,len(routes)):
    a = list(filter(None, re.split(regexPattern, routes.Items[i])))
    r.append([name for name in a if name.strip()])
routes.Items = r
```

```python
routes
```

```python
# Set the threshold for the support of Frequent itemset theory
support = 15
items = []
items = [item for items in routes.Items for item in items]
```

```python
from collections import Counter
itemCount = Counter(items)
```

```python
frequentItems = [name for name in itemCount if itemCount[name]>=support]
```

```python
len(frequentItems)
```

```python
a = {}
a = [list(filter(lambda i: i in frequentItems, items)) for items in routes.Items]
routes.Items = a
```

```python
from itertools import combinations
combination = [list(combinations(items, 2)) for items in routes.Items]
comb = [comb for combs in combination for comb in combs]
```

```python
pairCount = Counter(comb)
frequentPairs = [comb for comb in pairCount if pairCount[comb]>=support]
```

```python
frequentPairs = [sorted(fp) for fp in frequentPairs]
```

```python
frequentPairs
```

```python
popularTuples = frequentItems
popularTuples.extend(frequentPairs)
```

```python
popularTuples
```

```python
def matches(rid, order, popularTuples):
    itemList = []
    orderItems = routes.Items[(routes.RouteId == rid) & (routes.Order == order)].tolist()[0]
    for items in popularTuples:
        if isinstance(items,list):
            check = True
            for item in items:
                if item not in orderItems:
                    check = False
            if check:
                itemList.append(list(items))
        else:
            if items in orderItems:
                itemList.append(items)
    return itemList
    

def nextTuple(rid, order, popularTuples):
    if(not routes[(routes.RouteId == rid) & (routes.Order == order)].empty):
        ciao = matches(rid, order, popularTuples)
        return ciao
    else:
        return []

def searchNext(tuple, popularTuples):
    keys = [(routes.RouteId[row], routes.Order[row]) for row in range(0,len(routes)) if tuple in routes.Items[row]]
    newTuples = [nextTuple(key[0], key[1] + 1, popularTuples) for key in keys]
    return newTuples
```

```python
popularPatterns = [(pop, searchNext(pop, popularTuples)) for pop in popularTuples]
```

```python
popularPatterns
```

```python
itemList = []
for items in popularTuples:
    if isinstance(items,list):
        check = True
        for item in items:
            if item not in ['Milkshake', 'Strawberry-licorice']:
                check = False
        if check:
            itemList.append(list(items))
    else:
        if items in ['Milkshake', 'Strawberry-licorice']:
            itemList.append(items)
```
