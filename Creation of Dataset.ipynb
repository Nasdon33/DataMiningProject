```python
import pandas as pd

user_info=pd.read_csv('en.openfoodfacts.org.products.tsv',delimiter='\t',encoding='utf-8')
```

```python
import numpy as np

reduced = user_info.dropna(subset=['main_category'])
```

```python
list_product = np.unique(reduced['main_category_en'])
```

```python
pd.set_option("display.max_rows", 4000)
```

```python
final_product_list = pd.DataFrame(list_product[:1280])
final_product_list.columns = ['Name']
```

```python
final_product_list
```

```python
final_product_list['Popularity'] = np.random.random_sample(len(final_product_list))
```

```python
final_product_list
```

```python
final_product_list = final_product_list.drop([123,157,440,442])

# I obtain a list of product with a respective popularity, now I can generate a list of cities.
```

```python
cities = pd.read_csv('world-cities.csv',delimiter=',',encoding='utf-8')
```

```python
city_list = pd.DataFrame(cities.name)
city_list.columns = ['Name']
```

```python
city_list
```

```python
city_list["Export"] = ""
city_list["Import"] = ""
```

```python
# range of exported and imported for each city setted respectively at 3:20 and 10:50
for c in range(0,len(city_list)):
    num_exp = np.random.randint(3,20)
    num_imp = np.random.randint(10,50)
    exp = np.unique(final_product_list.sample(num_exp, weights=final_product_list.Popularity).Name)
    possible_imp_products = final_product_list[~final_product_list['Name'].isin(exp)]
    imp = np.unique(possible_imp_products.sample(num_imp, weights=possible_imp_products.Popularity).Name)
    city_list.loc[c,'Export'] = exp
    city_list.loc[c,'Import'] = imp
```

```python
# range of trip length setted at 3:15
length = np.random.randint(3,15)
trip = pd.DataFrame(columns=['IdCity','ExportableItems', 'PopularImport'])
items = []
for c in range(0,length):
    city = np.random.randint(0,len(city_list))
    trip.loc[c, 'IdCity'] = city_list.loc[city,'Name']
    items.extend(city_list.loc[city,'Export'])
    trip.loc[c, 'ExportableItems'] = items.copy()
    trip.loc[c, 'PopularImport'] = city_list.loc[city,'Import']
```

```python
trip
```

```python
# first i set a route id and a truck code for the first trip, then I will generate them with a loop.
rid = 0
tcd = 123
final_trip = pd.DataFrame(columns=['RouteId', 'Order', 'TruckCode', 'StartCity','EndCity', 'Items'])
for c in range(1,length):
    final_trip.loc[c-1, 'RouteId'] = rid
    final_trip.loc[c-1, 'Order'] = c-1
    final_trip.loc[c-1, 'TruckCode'] = tcd
    final_trip.loc[c-1, 'StartCity'] = trip.loc[c-1,'IdCity']
    final_trip.loc[c-1, 'EndCity'] = trip.loc[c,'IdCity']
    crev = length - c
    maxProbItems = list(set(trip.loc[crev+1, "PopularImport"]).intersection(trip.loc[crev, "ExportableItems"]))
    maxProbItems = maxProbItems * 5
    trip.loc[crev, "ExportableItems"].extend(maxProbItems)
    numItems = np.random.randint(5,20)
```
